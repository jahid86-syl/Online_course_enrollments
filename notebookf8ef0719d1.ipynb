```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/kaggle/input/dataset1/dataset_19_Online Course Enrollments.csv'
data = pd.read_csv(file_path)

# Group by 'Product' and sum the 'Revenue'
revenue_by_product = data.groupby('Product')['Revenue'].sum().reset_index()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(revenue_by_product['Product'], revenue_by_product['Revenue'], color='skyblue')

# Add labels and title
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.title('Total Revenue per Product')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
